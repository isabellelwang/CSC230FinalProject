```python
import spacy
import pandas as pd
pd.set_option('display.max_colwidth', None)
nlp = spacy.blank("en")
nlp.add_pipe("sentencizer")
```

```python
df = pd.read_csv("Testimonies.csv")
```

```python
df.columns
```

```python
## empty list for the results
results = []

## loops through data one row at a time (ignoring row index because we dont care)
for _, row in df.iterrows():
    fname = row["fname"]
    lname = row["lname"]
    text = row["testimony"]

    if pd.isna(text):
        continue

    ## makes the names full (so we have first and last names)
    full_name = f"{fname} {lname}"

    ## sends text to spaCy WOOOoooo
    doc = nlp(str(text))

    ## checks if the sentence has husband or husbands (makes it all lowercase to avoid issues there)
    for sent in doc.sents:
        s = sent.text.lower()
        if "husband" in s or "husbands" in s:
            results.append({
                "name": full_name,
                "sentence": sent.text.strip()})

##just testing if it works lol 
results[:10]

## move results to another csv file
pd.DataFrame(results).to_csv("husband_sentences.csv", index = False)
```
