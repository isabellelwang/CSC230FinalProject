```python
import pandas as pd
import sqlite3

# Load CSV file
df = pd.read_csv('test.csv')
print(df.head(10))
df.rename(columns={'story arc': 'story_arc',"husband support": "husband_support" }, inplace=True)
print(df.head(10))

# Create a connection to db file
conn = sqlite3.connect('test.db')

#
df.to_sql('test_table', conn, if_exists='replace', index=False)

# Close the connection
conn.close()
```

```python
DB_PATH = "test.db"

conn = sqlite3.connect(DB_PATH)
conn.execute("PRAGMA foreign_keys = ON")

def run_sql(sql: str):
    try:
        cur = conn.execute(sql)
        if sql.lstrip().upper().startswith(("SELECT", "PRAGMA")):
            rows = cur.fetchall()
            print(rows)
        else:
            conn.commit()
            print("OK")
    except sqlite3.Error as err:
        print(f"SQLite error: {err}")
```

```python
run_sql("""SELECT name FROM sqlite_master WHERE type='table'""")
```

```python
# Shows all rows and all columns
run_sql("SELECT * FROM my_table")
```

```python
run_sql(
    """
    SELECT COUNT(husband_support)
    FROM my_table
    WHERE husband_support = "Not_Supportive"
    """
)
```
